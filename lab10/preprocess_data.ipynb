```python
import xml, os
import xml.etree.ElementTree as ET
import pprint
```

```python
def read_parse_files(source_path="dane/"):
    files = os.listdir(source_path)
    files_path = [source_path + file for file in files]
    files_content = []
    Y = []
    for file_name in files_path:
        with open(file_name) as file:
            print (file_name)
            tree = ET.parse(file)
            root = tree.getroot()
            for child in root:
                if child.tag == 'chunkList':
                    sentences_in_file = []
                    for sentence in child:
                        new_sentence = []
                        if sentence.tag == 'chunk':
                            for token in sentence:
                                if token.tag == 'tok':
                                    for tok in token:
                                        if tok.tag == 'orth':
                                            new_sentence.append(tok.text)
                        sentences_in_file.append(new_sentence)
            files_content.append(sentences_in_file)
            if "F" in file_name:
                Y.append(0)
            else:
                Y.append(1)
    return files_content, Y
```

```python
content, Y = read_parse_files()
```

```python
final_content = []
for file in content:
    merged = []
    for blah in file:
        for item in blah:
            merged.append(item)
    final_content.append(merged)
```

```python
print(final_content)
```

```python
lines = []
for idx, content in enumerate(final_content):
    joined = " ".join(content)
    y = Y[idx]
    y_str = "__label__" + str(y) + " "
    content = y_str + joined
    lines.append(content)

with open("fake_reviews.txt", "w") as file:
    for line in lines:
        file.write(line + "\n")
```
